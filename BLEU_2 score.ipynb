```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.translate.bleu_score import sentence_bleu
```

```python
def list_of_list(lst):
    
    return [[el] for el in lst if el!='']


def extract_subs(text):
    places = []

# open file and read the content in a list
    with open(text, 'r') as filehandle:
        for line in filehandle:
        # remove linebreak which is the last character of the string
            currentPlace = line[:-1]

        # add item to the list
            if currentPlace!= '':
                places.append(currentPlace)
    places_2 = list_of_list(places)
    return places
```

```python
def get_tfidf(str1, str2):
    sub_1_string = ''.join(i for i in str1)
    sub_2_string = ''.join(i for i in str2)
    tfidf = TfidfVectorizer()
    vecs = tfidf.fit_transform([sub_1_string, sub_2_string])
    corr_matrix = ((vecs * vecs.T).A)
    return corr_matrix
```

```python
def get_bleu_score(str1, str2):
    sub_1_string = ''.join(i for i in str1)
    sub_2_string = ''.join(i for i in str2)
    return sentence_bleu(sub_1_string, sub_2_string)
```
